````python
def strange_matrix(n: int, first: int, delta: int) -> np.matrix:
    result_matrix = np.eye(n)
    nums = np.array([i for i in range(first, n * delta + 1, delta)])
    result_matrix *= nums
    return result_matrix
print(strange_matrix(5, 1, 1))
````

### 1) Какие структуры данных используются в Numpy? В чем их отличие от списков Python?

| Отличие             | `ndarray` (NumPy)                            | `list` (Python)                            |
|---------------------|----------------------------------------------|---------------------------------------------|
| **Тип данных**      | Однородные (все одного типа)                | Разные типы в одном списке                  |
| **Производительность** | Быстрее (работает на C)                  | Медленнее (интерпретируемый Python)         |
| **Память**          | Использует меньше памяти                    | Использует больше памяти                    |
| **Операции**        | Векторные (без циклов)                      | Требуются циклы или map                     |
| **Многомерность**   | Поддерживается (`.shape`, `.ndim`)          | Только вложенные списки                     |
| **Инструменты**     | Много встроенных функций                    | Почти нет встроенных                        |

### 3) Какие способы предлагает Numpy для извлечения данных из массивов?

В NumPy для извлечения данных из массивов используются следующие способы:

1. Индексация:

    * Для одномерных массивов: array[index]

    * Для многомерных: array[row, col]

2. Срезы (slicing):

    * Одномерные: array[start:end:step]

    * Многомерные: array[start_row:end_row, start_col:end_col]

3. Маски (boolean indexing):

    Пример: array[array > 0] — извлекает элементы, удовлетворяющие условию.

4. Функции:

    * np.take() — извлекает элементы по индексам.

    * np.where() — находит индексы, удовлетворяющие условию.

### 4) Что такое векторизация кода и почему это ускоряет работу программ?

Векторизация кода — это процесс преобразования операций с элементами массива в операции, которые могут быть выполнены за один раз над всей коллекцией данных, без использования явных циклов.

Почему это ускоряет работу:
1. Меньше циклов: Векторизация заменяет медленные циклы Python на более быстрые операции, реализованные на C или Fortran в NumPy.

2. Параллельное выполнение: Векторные операции могут быть оптимизированы для работы с несколькими данными одновременно (на уровне процессора).

3. Меньше накладных расходов: Прямые операции с массивами требуют меньше времени на интерпретацию, чем многократные циклы.

### 5) Какие виды матричных операций реализованы в Numpy?

### Виды матричных операций в NumPy:

1. **Операции сложения, вычитания и умножения**:
   - `+`, `-`, `*` — поэлементные операции (для матриц, элементов массива).

2. **Умножение матриц**:
   - `np.dot(A, B)` или `A @ B` — скалярное произведение (умножение матриц).
   - `np.matmul(A, B)` — для умножения матриц (аналогично `@`).

3. **Транспонирование**:
   - `np.transpose(A)` или `A.T` — транспонирование матрицы.

4. **Обратная матрица**:
   - `np.linalg.inv(A)` — находит обратную матрицу.

5. **Определитель**:
   - `np.linalg.det(A)` — вычисляет определитель матрицы.

6. **Собственные значения и собственные векторы**:
   - `np.linalg.eig(A)` — находит собственные значения и векторы матрицы.

7. **Детерминант, след, ранг**:
   - `np.trace(A)` — след матрицы.
   - `np.linalg.matrix_rank(A)` — ранг матрицы.

8. **Разложение матриц**:
   - `np.linalg.svd(A)` — сингулярное разложение.
   - `np.linalg.eig()` — для собственных значений и векторов.


### 6) Какие функции используются для преобразования формы, размера и соединения массивов?

### Функции для преобразования формы, размера и соединения массивов в NumPy:

#### Преобразование формы и размера:
1. **`np.reshape()`** — изменяет форму массива, не изменяя его данных.
   - Пример: `np.reshape(array, (new_shape))`
   
2. **`np.resize()`** — изменяет размер массива, может добавлять новые элементы (повторяя данные).
   - Пример: `np.resize(array, (new_size))`

3. **`np.flatten()`** — преобразует многомерный массив в одномерный.
   - Пример: `array.flatten()`

4. **`np.ravel()`** — возвращает одномерный массив, изменяя существующую форму (если возможно).
   - Пример: `array.ravel()`

5. **`np.transpose()`** — транспонирует массив.
   - Пример: `np.transpose(array)`

6. **`np.newaxis`** — добавляет новую ось в массив (для расширения размерности).
   - Пример: `array[:, np.newaxis]`

#### Соединение массивов:
1. **`np.concatenate()`** — соединяет массивы вдоль указанной оси.
   - Пример: `np.concatenate((array1, array2), axis=0)`

2. **`np.stack()`** — соединяет массивы по новой оси.
   - Пример: `np.stack((array1, array2), axis=0)`

3. **`np.hstack()`** — горизонтальное соединение (по оси 1).
   - Пример: `np.hstack((array1, array2))`

4. **`np.vstack()`** — вертикальное соединение (по оси 0).
   - Пример: `np.vstack((array1, array2))`

5. **`np.column_stack()`** — соединяет массивы по столбцам (по оси 1).
   - Пример: `np.column_stack((array1, array2))`

6. **`np.row_stack()`** — соединяет массивы по строкам (по оси 0).
   - Пример: `np.row_stack((array1, array2))`


### 7) Какие две главные структуры данных используются в pandas? В чем их отличие?

### Две главные структуры данных в pandas:

1. **Series**:
   - Это одномерный массив, который может содержать любые данные (например, числа, строки). Он имеет индекс, который позволяет обращаться к значениям с помощью меток.
   - По сути, Series — это столбец таблицы.

2. **DataFrame**:
   - Это двумерная структура данных, которая представляет собой таблицу, состоящую из строк и столбцов. Каждый столбец в DataFrame является объектом Series.
   - DataFrame можно рассматривать как коллекцию Series, где каждая колонка имеет метки индекса.

### Основные отличия:
- **Series**: одномерная структура (один столбец данных).
- **DataFrame**: двумерная структура (таблица с несколькими столбцами и строками).


### 8) Как происходит объединение двух таблиц в pandas?

### Объединение двух таблиц в pandas

В pandas существует несколько способов объединения таблиц (DataFrame). Основные методы:

1. **`pd.concat()`**:
   - Используется для объединения DataFrame вдоль заданной оси (по строкам или по столбцам).

2. **`pd.merge()`**:
   - Применяется для объединения таблиц на основе общих столбцов или индексов, аналогично SQL-операциям `JOIN`.
   - Может быть использован для объединения по индексам или конкретным столбцам.

3. **`join()`**:
   - Метод для объединения таблиц по индексам или столбцам, аналогичен `pd.merge()`, но с ограничениями.
   
### Основные параметры:
- **`on`** — имя столбца для объединения.
- **`left_on`, `right_on`** — указание столбцов для объединения из разных таблиц.
- **`how`** — тип объединения (`'inner'`, `'outer'`, `'left'`, `'right'`).

### Отличия:
- **`concat()`** используется для простого объединения данных без учета ключей.
- **`merge()`** и **`join()`** более гибкие и позволяют объединять данные на основе ключей или индексов.


### 9) Зачем нужны и как работают индексы в pandas.

### Индексы в pandas

**Зачем нужны:**
- Уникальная идентификация строк и столбцов.
- Ускорение поиска, сортировки и фильтрации данных.
- Эффективная работа с временными рядами.
- Использование для объединения и слияния данных.

**Как работают:**
- По умолчанию pandas создает числовой индекс (0, 1, 2, …).
- Можно задавать собственные индексы с помощью `set_index()`.
- Поддержка многократных индексов (MultiIndex).
- Для доступа используется `.loc[]` (по метке) и `.iloc[]` (по позиции).

**Операции с индексами:**
- Сброс индекса: `df.reset_index()`.
- Проверка индекса: `df.index`.


### 10) Построение каких основных видов графиков используется при анализе данных в машинном обучении?

### Основные виды графиков при анализе данных в машинном обучении

1. **Гистограмма (Histogram)**:
   - Для визуализации распределения данных.
   - Показывает частоту значений переменных.

2. **Диаграмма рассеяния (Scatter Plot)**:
   - Для анализа взаимосвязи между двумя переменными.
   - Визуализирует зависимость между признаками.

3. **Ядерная оценка плотности (KDE Plot)**:
   - Для визуализации распределения данных в виде сглаженной кривой.
   - Используется для оценки плотности вероятности.

4. **Ящиковая диаграмма (Box Plot)**:
   - Для выявления выбросов и визуализации статистических характеристик (медиана, квартиль).

5. **Матрица корреляций (Correlation Heatmap)**:
   - Для оценки зависимости между признаками с помощью цветовой карты.

6. **Линейный график (Line Plot)**:
   - Для визуализации изменений переменной во времени или по порядку.

7. **Парные графики (Pair Plot)**:
   - Для визуализации попарных отношений между несколькими признаками.

8. **Круговая диаграмма (Pie Chart)**:
   - Для отображения доли каждого компонента в общем наборе данных.

Эти графики помогают понять структуру данных, выявить закономерности и аномалии, а также подготовить данные для дальнейшего анализа в моделях машинного обучения.

### 11) В чём разница между библиотеками matplotlib и seaborn? Каковы преимущества каждой из них?

### Разница между matplotlib и seaborn

1. **matplotlib**:
   - **Основное назначение**: Базовая библиотека для создания статических, анимационных и интерактивных графиков.
   - **Преимущества**:
     - Полная гибкость в создании графиков (можно настроить все аспекты).
     - Широкие возможности для создания различных типов графиков (линии, гистограммы, scatter plot и т.д.).
     - Широко используется и поддерживается в научном сообществе.
   - **Недостатки**:
     - Требует больше кода для настройки графиков.
     - Для некоторых визуализаций нужно много ручной настройки.

2. **seaborn**:
   - **Основное назначение**: Построение статистических графиков на основе matplotlib с более простым и высокоуровневым API.
   - **Преимущества**:
     - Простота в использовании для статистических графиков.
     - Стиль и визуальные улучшения по умолчанию (красивые графики с минимальными усилиями).
     - Интеграция с pandas — легко работать с DataFrame.
     - Отлично подходит для визуализации корреляций, распределений и других статистических данных.
   - **Недостатки**:
     - Менее гибкий, чем matplotlib (например, для кастомизации графиков).

### Когда использовать:

- **seaborn**: Когда нужно быстро построить статистические графики с красивым оформлением, особенно при работе с pandas.

### 15) Какие основные типы графиков реализованы в matplotlib? Что изображается на ящиковой диаграмме?

### Основные типы графиков в matplotlib

1. **Линейный график (Line Plot)**:
   - Отображает зависимость между переменными.
   - Идеален для отображения временных рядов или трендов.

2. **Гистограмма (Histogram)**:
   - Показывает распределение данных, деля их на интервалы (бини).
   - Используется для оценки частотного распределения.

3. **Диаграмма рассеяния (Scatter Plot)**:
   - Отображает взаимосвязь между двумя переменными.
   - Полезен для выявления корреляции.

4. **Ядерная оценка плотности (KDE Plot)**:
   - Гладкое представление распределения данных.
   - Используется для оценки плотности вероятности.

5. **Ящиковая диаграмма (Box Plot)**:
   - Показывает распределение данных с использованием квартилей.
   - Выявляет медиану, межквартильный размах и выбросы.

6. **Круговая диаграмма (Pie Chart)**:
   - Показ доли каждой категории от общего набора.
   - Часто используется для представления пропорций.

7. **Стадионная диаграмма (Area Plot)**:
   - Отображает кумулятивные значения через области, заполненные цветом.

8. **Точечная диаграмма (Stem Plot)**:
   - Показывает данные с помощью стеблей и точек.

9. **Барграф (Bar Plot)**:
   - Для сравнения категориальных данных с помощью столбцов.

### Ящиковая диаграмма (Box Plot)

На **ящиковой диаграмме** изображаются следующие компоненты:
- **Медиана (Median)**: Центральное значение набора данных (линия в коробке).
- **Квартильные значения**:
  - **Q1 (25-й процентиль)**: Нижняя граница коробки.
  - **Q3 (75-й процентиль)**: Верхняя граница коробки.
- **Межквартильный размах (IQR)**: Разница между Q3 и Q1, показывающая распределение центральной части данных.
- **Выбросы**: Точки, которые выходят за пределы диапазона (обычно 1.5 * IQR).
