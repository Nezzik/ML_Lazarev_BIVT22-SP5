Парамет how у merge аналогичен способам объединения из sql

INNER JOIN — возвращает только те строки, у которых есть совпадения в обеих таблицах.

LEFT JOIN (или LEFT OUTER JOIN) — возвращает все строки из левой таблицы и совпадающие строки из правой. 

RIGHT JOIN (или RIGHT OUTER JOIN) — аналогично, но с приоритетом правой таблицы.

FULL OUTER JOIN — возвращает все строки из обеих таблиц, совпадающие и несовпадающие.

### 5. Изучите документацию pandas в части описания методов merge и join.

#### Основные отличия:

| Характеристика      | `merge()`                                                                      | `join()`                                |
| ------------------- | ------------------------------------------------------------------------------ | --------------------------------------- |
| **Тип объединения** | По столбцам                                                                    | По индексам или указанным столбцам      |
| **Гибкость**        | Высокая (можно объединять по разным столбцам)                                  | Ограниченная (по умолчанию по индексам) |
| **Тип объединения** | Поддерживает все типы объединений (`inner`, `outer`, `left`, `right`, `cross`) | Поддерживает все типы объединений       |
| **Использование**   | Когда нужно объединить по столбцам                                             | Когда нужно объединить по индексам      |

# Контрольные вопросы

**Какие ошибки могут произойти при горизонтальной интеграции данных?**  
Несовпадение индексов, дублирование строк, несогласованные значения ключей, разные шкалы признаков.

**Какими средствами можно проверить совпадение шкал измерения атрибута в разных датасетах?**  
Гистограммы (`df.hist()`), плотность (`sns.kdeplot`), описательная статистика (`df.describe()`), нормализация (`StandardScaler` / `MinMaxScaler`).

**Какие ошибки могут произойти при вертикальной интеграции данных?**  
Разное количество столбцов, несовпадение названий столбцов, разные типы данных, дубликаты строк.

**Чем отличаются методы merge и join в библиотеке pandas?**  
`merge` более гибкий, работает с любыми колонками как ключами, `join` проще — работает по индексу по умолчанию.

**Какие бывают виды соединения таблиц и в каких случая их нужно применять? Приведите по два примера на каждый вид объединения.**

`inner` — оставить только пересечения:
- Клиенты с заказами
- Пользователи, у которых есть метки активности

`left` — сохранить все из левой:
- Все товары + данные о продажах
- Все заказы + статусы (если есть)

`right` — сохранить все из правой:
- Все активные агенты + их обращения
- Все действующие продукты + сведения о цене

`outer` — полное объединение:
- Полный список клиентов + все жалобы
- Все сотрудники из двух департаментов
